### create a new dataframe with only the relevant values
- there was too much NaN values and the only columns that did not have any NaN values are CPI and Global Insight Country Risk Ratings.

```python
df2 = pd.DataFrame(df1, columns=['Country', 'Region', 'Corruption Perceptions Index (CPI)', 'Global Insight Country Risk Ratings'])
df2
```

### Simplify Columns

```python
df2['GI'] = df2['Global Insight Country Risk Ratings']
df2['CPI'] = df2['Corruption Perceptions Index (CPI)']
df2 = df2.drop(['Global Insight Country Risk Ratings', 'Corruption Perceptions Index (CPI)'], axis=1)
df2
```

## Create a function to calculate average between two values and apply it using lamda function
### Column 'ULT' is the average between Global Insight Country Risk Ratings and Corruption Perceptions Index

```python
def calculate(gi, cpi):
    result = (3 * gi + 2 * cpi) / 5
    return result

df2['ULT'] = df2[['GI', 'CPI']].apply(lambda x: calculate(x["GI"], x["CPI"]), axis=1)
df2
```

### Top 10 countries by ULT

```python
df2.sort_values('ULT', ascending=False).head(10)
```

### Worst 10 countries by ULT

```python
df2.sort_values('ULT', ascending=True).head(10)
```

## Find the average ULT Score by Region

```python
regions = df2.drop(['Country', 'GI', 'CPI'], axis=1)
regions1 = regions.groupby('Region')
regions2 = regions1.mean().sort_values('ULT', ascending=False)
regions2
```

## save the dataframes as a new csv file

```python
df3 = df2.sort_values('ULT', ascending=False)
df3.reset_index(drop=True)
df3.to_csv('my_analysis.csv')
regions2.to_csv('regions.csv')
```
